#**Filters and Sorting**

```python
dollarizer = lambda x: float(x[1:-1])
df_chipotle.item_price = df_chipotle.item_price.apply(dollarizer)
df_chipotle
```

```python
df_chipotle[df_chipotle.item_price > 10]
```

```python
df_chipotle[(df_chipotle.item_price > 10) & (df_chipotle.item_name == 'Chicken Bowl')]
```

```python
df_chipotle[df_chipotle.item_name == 'Chicken Bowl']
```

```python
df_chipotle[df_chipotle.item_name == 'Chicken Bowl'].value_counts().sum()
```

```python
df_chipotle_dup = df_chipotle.drop_duplicates(['item_name', 'quantity'])
```

```python
df_chipotle_dup
```

```python
df_chipotle_once = df_chipotle_dup[(df_chipotle_dup.quantity==1)]
```

```python
df_chipotle_once
```

```python
df_price_per_item = df_chipotle_once[['item_name', 'item_price']]
```

```python
df_price_per_item
```

```python
df_price_per_item.sort_values(by='item_price', ascending=True)
```

```python
import pandas as pd


users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',
                      sep='|', index_col='user_id')
users.head()
```

```python
users
```

```python
users.groupby('occupation').age.mean()
```

```python
users.groupby('occupation').age.agg(['min','max'])
```

```python
users.groupby(['occupation', 'gender']).age.mean()
```

```python
users.groupby(['occupation', 'gender']).age.agg(['min', 'max', 'mean'])
```
