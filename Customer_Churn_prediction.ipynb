```python
account_length_distribution = df['Account length'].value_counts().reset_index()
account_length_distribution.columns = ['Account Length', 'Customer Count']
print("\nAccount Length Analysis:")
print(account_length_distribution)
```

```python
area_code_distribution = df['Area code'].value_counts().reset_index()
area_code_distribution.columns = ['Area Code', 'Customer Count']
print("\nArea Code Distribution:")
print(area_code_distribution)
```

```python
international_plan_churn = df.groupby(['International plan', 'Churn']).size().reset_index(name='Customer Count')
print("\nInternational Plan and Churn Analysis:")
print(international_plan_churn)
```

```python
voice_mail_plan_churn = df.groupby(['Voice mail plan', 'Churn']).size().reset_index(name='Customer Count')
print("\nVoice Mail Plan and Churn Analysis:")
print(voice_mail_plan_churn)
```

```python
num_vmail_messages_distribution = df['Number vmail messages'].value_counts().reset_index()
num_vmail_messages_distribution.columns = ['Number of Voice Mail Messages', 'Customer Count']
print("\nNumber of Voice Mail Messages Analysis:")
print(num_vmail_messages_distribution)
```

```python
customer_service_calls_distribution = df['Customer service calls'].value_counts().reset_index()
customer_service_calls_distribution.columns = ['Number of Customer Service Calls', 'Customer Count']
print("\nCustomer Service Calls Analysis:")
print(customer_service_calls_distribution)
```
